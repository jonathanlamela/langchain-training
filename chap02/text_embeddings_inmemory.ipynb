```python
from langchain_community.document_loaders import TextLoader
```

```python
loader = TextLoader("./genesi.txt")
```

```python
docs = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
```

```python
splitted_docs = splitter.split_documents(docs)
```

```python
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings_model = OllamaEmbeddings(model="embeddinggemma:latest")
```

```python
from langchain_core.vectorstores import InMemoryVectorStore
```

```python
vector_store = InMemoryVectorStore(embeddings_model)
```

```python
vector_store.add_documents(splitted_docs)
```

```python
vector_store.similarity_search(query="inizialmente Dio creò la terra e i cieli",k=4)
```
